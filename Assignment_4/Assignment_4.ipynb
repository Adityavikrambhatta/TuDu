#Task 2: Read File in Different Ways

```python
# Using .read()

file = open("sales_data.txt", "r")
print(file.read())
file.close()

# Using .readline()

with open("sales_data.txt", "r") as file:
  print(file.readline())
file.close()

# Using .readlines()
list_int = []

file = open("sales_data.txt", "r")
list_int = file.readlines()
file.close()

list_int = [int(i) for i in list_int]
```

# Task 4: Generate Summary Report from File

```python
def total_sales(sales_list):
  total_sales = 0
  for sale in sales_list:
    total_sales += sale
  return total_sales

def average_sales(sales_list):
  average_sales = total_sales(sales_list) / len(sales_list)
  return average_sales

def highest_sale(sales_list):
  highest_sale = 0
  for sale in sales_list:
    if sale > highest_sale:
      highest_sale = sale
  return highest_sale

def lowest_sale(sales_list):
  lowest_sale = float('inf')
  for sale in sales_list:
    if sale < lowest_sale:
      lowest_sale = sale
  return lowest_sale

def read_file(file):
  file = open("sales_data.txt", "r")
  sales_list = file.readlines()
  file.close()
  sales_list = [int(i) for i in list_int]
  return sales_list

sales_list = read_file("sales_data.txt")
print("Total Sales: ", total_sales(sales_list))
print("Highest Sale: ", highest_sale(sales_list))
print("Lowest Sale: ", lowest_sale(sales_list))
print("Average Sales: ", average_sales(sales_list))

```

# Task 5: Create Product Info File (User Input)

```python



products = []
three = 3
for i in range(three):
  input_product_name = input(f"Enter product {three-i} name: ")
  input_product_price = input(f"Enter the {three-i} product price: ")
  products.append(f"{input_product_name}|{input_product_price}")

file = open("products.txt", "w")
for product in products:
  file.write(str(product)+ "\n")
file.close()

file = open("products.txt", "r")
print(file.read())
file.close()

```

# Task 7: Mini Project - Export Discounted Prices

```python

prices = {
    "Mouse" : 500,
    "Keyboard" : 800,
    "Monitor" : 7000,
    "Pendrive" : 400,
    "Camera": 5000,
}

discount_percentage = int(input("Enter the discount percentage: "))
discounted_prices = {}
for product, price in prices.items():
  discounted_prices[product] = price * (1 - discount_percentage / 100)

file = open("discount_report.txt", "w")
file.write("Product | Original Price | Discounted Price\n")

print(prices)
for product, price in prices.items() :
  file.write(f"{product} | {price} | {discounted_prices[product]}\n")

file.close()

file = open("discount_report.txt", "r")
all_prices = file.read()
file.close()

total_discounted_price = 0
count = 0
for product, price in discounted_prices.items():
  total_discounted_price += price
  count += 1

print(all_prices)
print("Total Items: ", count)
print(f"Average Discounted Prices: {total_discounted_price/count}", )


```
