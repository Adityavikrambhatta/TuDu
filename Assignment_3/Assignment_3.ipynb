## Task 1 - Basic Function : PriceAfter Discount

```python
def apply_discount(price, discount_percentage = 5):
  """
  The function calculates the discount price and retuns the discounted price
  If the discount percentage is not provided, it defaults to 5%.
  """
  if discount_percentage > 60 :
    discount_percentage = 60

  discount_amount = price * (discount_percentage / 100)
  price_after_discount = price - discount_amount

  return price_after_discount
```

```python
print(apply_discount(1000,10))
print(apply_discount(500))

# Optional boundary condition check
print(apply_discount(1000, 75))
```

## Task 2 - Recursive Function : Factorial Utility

```python
def factorial(n) :
  if n == 1 or n == 0:
    return 1
  elif n < 0:
    return "Invalid Input"
  else :
    return n * factorial(n-1)
```

```python
print(factorial(5))
print(factorial(0))
print(factorial(-3))
```

## Task 3 - Lambda Function: GST Calculator

```python
gst = lambda price : price + (price * 0.18)
print(gst(100))
```

```python
# Extra (Optional)
# I am trying to calculate GST + discount to calculate the final price
discount_percent= 21
discount = lambda price : (price * discount_percent/100)
# GST + discount together
final_price = gst(100) + discount(100)
print(final_price)
```

## Task 4 - Using map(): Apply GST to List of Prices

```python
prices = [100, 250, 400, 1200, 50]
gst_prices = list(map(gst, prices))
print(f"Price after GST : {gst_prices}")
```

## Task 5 - Using filter() : Filter Expensive Products

```python
prices = [100, 250, 400, 1200, 50, 2000, 850]
expensive_products = list(filter(lambda price : price > 500, prices))
print(f"Expensive Products : {expensive_products}")
```

## Task 6 - Combined Utility Functions

```python
def process_prices(prices):
  discounted_prices = list(map( lambda price : price - (price * 0.10) , prices))
  prices_greater_than_300 = list(filter(lambda price : price > 300, prices))
  return discounted_prices, prices_greater_than_300
```

```python
discount_price, filtered_price = process_prices([100, 500, 900, 50, 750])
print(f"Discounted Prices: {discount_price}")
print(f"Filtered Prices (prices > 300): {filtered_price}")
```

## Task 7 - Mini Problem: Menu Using Functions

```python
import builtins
def add_price(price_list, new_price):
  price_list.append(new_price)
  return price_list
def max_price(price_list):
  max = 0
  for val in price_list:
    if val > max:
      max = val
  return max

averages = lambda price_list : sum(price_list)/len(price_list)
max_val = 0

def menu(price_list):
  while True:
    print("Menu : ")
    print("1. Add Price")
    print("2. Average Price")
    print("3. Max Price")
    print("q. Quit")

    choice = input("Enter your choice (1/2/3/q) : ")

    if choice == '1':
      try:
        new_price = int(input("Enter the new price : "))
        price_list = add_price(price_list, new_price)
      except ValueError:
        print("Invalid price. Please enter a number.")
    elif choice == '2':
      print(f"Average Price : {averages(price_list)}")
    elif choice == '3':
      print(f"Max Price : {max_price(price_list)}")
    elif choice == 'q':
      print("Thank You!")
      break
    else:
      print("Invalid Choice")
```
