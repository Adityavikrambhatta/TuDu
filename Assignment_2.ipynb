## Task 1: Discount Rules (if/elif/else)

1. Write a program that reads an integer order_amount from the user using input()

## Task 2: Process Multiple Orders (for loop)

1. Given a list of order amounts: orders = [1200, 2500, 800, 1750, 3000], use a for loop to apply the discount rules from Task 1 to each order and print a summary table showing: order_amount -> discount% -> final_amount.
2. Also compute and print the total revenue after disocunts.
- Extra (optinal): Print the number of orders that receiveda discount (discount > 0)


```python
orders = [1200, 2500, 800, 1750, 3000]
total_revenue_after_discount = 0
num_orders_with_discount = 0

print("\n--- Order Summary ---")
print(f"{'Order Amount':<15} {'Discount %':<15} {'Final Amount':<15}")
print(f"{'='*15:<15} {'='*15:<15} {'='*15:<15}")

for order_amount in orders:
    discount_percentage = 0
    if order_amount >= 2000:
        discount_percentage = 0.15 # 15% discount
    elif order_amount >= 1500:
        discount_percentage = 0.10 # 10% discount
    elif order_amount >= 1000:
        discount_percentage = 0.07 # 7% discount

    discount_amount = order_amount * discount_percentage
    final_amount = order_amount - discount_amount

    if discount_percentage > 0:
        num_orders_with_discount += 1

    total_revenue_after_discount += final_amount
    print(f"{order_amount:<15.2f} {discount_percentage * 100:<14.0f}% {final_amount:<15.2f}")

print(f"\nTotal Revenue After Discounts: ${total_revenue_after_discount:.2f}")
print(f"Number of orders that received a discount: {num_orders_with_discount}")
```

## Task 3: User Menu (while loop + break/continue)
1. Implement a simple while-loop driven menu that keeps asking the user to choose an action until they press q or quit.
Menu options:
- 1-- Add order amount to a running list (use input())
- 2-- Show all orders and totals after applying discounts
- q --Quit
2. Use continue to re-show the menu after invalid input and break to exit on q.

Constraints : Keep the loop logic simple and use only lists and loops (no functions required).

```python
all_orders = []

while True:
    print("\n--- User Menu ---")
    print("1 -- Add order amount")
    print("2 -- Show all orders and totals")
    print("q -- Quit")
    choice = input("Enter your choice: ").lower()

    if choice == '1':
        try:
            order_amount_str = input("Enter the order amount: ")
            order_amount = float(order_amount_str)
            if order_amount < 0:
                print("Error: Order amount cannot be negative.")
                continue
            all_orders.append(order_amount)
            print(f"Order of ${order_amount:.2f} added.")
        except ValueError:
            print("Error: Invalid input. Please enter a numeric value.")
            continue
    elif choice == '2':
        if not all_orders:
            print("No orders to display yet.")
            continue

        total_revenue_after_discount = 0
        num_orders_with_discount = 0

        print("\n--- All Orders Summary ---")
        print(f"{'Order Amount':<15} {'Discount %':<15} {'Final Amount':<15}")
        print(f"{'='*15:<15} {'='*15:<15} {'='*15:<15}")

        for order_amount_item in all_orders:
            discount_percentage = 0
            if order_amount_item >= 2000:
                discount_percentage = 0.15 # 15% discount
            elif order_amount_item >= 1500:
                discount_percentage = 0.10 # 10% discount
            elif order_amount_item >= 1000:
                discount_percentage = 0.07 # 7% discount

            discount_amount = order_amount_item * discount_percentage
            final_amount = order_amount_item - discount_amount

            if discount_percentage > 0:
                num_orders_with_discount += 1

            total_revenue_after_discount += final_amount
            print(f"{order_amount_item:<15.2f} {discount_percentage * 100:<14.0f}% {final_amount:<15.2f}")

        print(f"\nTotal Revenue Across All Orders (After Discounts): ${total_revenue_after_discount:.2f}")
        print(f"Number of orders that received a discount: {num_orders_with_discount}")
    elif choice == 'q':
        print("Exiting program. Goodbye!")
        break
    else:
        print("Invalid choice. Please select 1, 2, or q.")
        continue
```

## Task 5: Loop Control with Conditions (break and continue)

1. Given a list of daily sales: daily = [200, 150, 0, 400, 50, -1, 300], iterate through it with a for loop and:
- If a day's value is -1, treat it as corrupted data and break the loop (stop processing).
- If a day's value is 0, treat it as a day with no slaes and continue(skip adding to revenue).
- For valid positive sales, add to total_sales and print the running total.
2. Print the final total after the loop completes (or stop due to corrupted data).


```python
daily_sales = [200, 150, 0, 400, 50, -1, 300]
total_sales = 0

print("\n--- Daily Sales Processing ---")

for sales_value in daily_sales:
    if sales_value == -1:
        print("Corrupted data (-1) encountered. Stopping processing.")
        break # Stop the loop entirely
    elif sales_value == 0:
        print(f"Day with 0 sales. Skipping. Current total: ${total_sales:.2f}")
        continue # Skip to the next iteration
    elif sales_value > 0:
        total_sales += sales_value
        print(f"Added sales of ${sales_value:.2f}. Running total: ${total_sales:.2f}")
    else:
        # Handle any unexpected negative values (not -1)
        print(f"Warning: Encountered unexpected negative sales value: ${sales_value:.2f}. Skipping.")

print(f"\nFinal Total Sales: ${total_sales:.2f}")
```
