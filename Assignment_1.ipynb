### Task 4: Combined Operations

1. Using the products list and price_dict, create a list odtuples named catalog where each tuple is (product_name,price,category).

```python
# Iniitializing lists
product_list = [ "adidas_original_sneakers", "jordans_sneakers", "commet_sneakers", "arrow_trousers", "levis_jeans", "timex_watch" ]
product_price = [10999, 40000, 5000, 2000, 4000, 4500]
product_category = [ "shoes", "shoes", "shoes", "pants", "pants", "watch"]
catalog = list()

for index, value in enumerate(product_list):
  catalog.append((value, product_price[index], product_category[index]))
catalog
```

2. From catalog, create a new dictionary category_to_products that maps each category to a list of products in that category.

```python
category_to_products = dict()

for product in catalog:
  if product[2] in category_to_products:
    category_to_products[product[2]].append(product[0])
  else:
    category_to_products[product[2]] = [product[0]]
category_to_products
```

3. Print all products that belong to the category thta has the maximum number of Products.

```python
max_category = max(category_to_products, key=lambda x: len(category_to_products[x]))
print(max_category)
print(category_to_products[max_category])
```
